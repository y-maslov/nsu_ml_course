```python
"""1.Сортировка в обратном порядке
Пользователь сначала вводит количество чисел, а затем вводит сами числа, ваша задача
отсортировать введенные пользователем числа и вывести их в обратном порядке."""
n = int(input())
a = [int(input()) for i in range(n)]
a.sort()
for i in range(n):
    print(a[n - i - 1])

```

```python
"""2.Фильтр
Напишите программу, которая проводит первичную обработку неких сложных и
глючных логов. Нужно удалить сочетание «%%» в начале некоторых строк и удалить
строки, начинающиеся с «####».
Пример:
ввод:
3
SVO TRS 29481292
%%LJPZ DME 11113283675
###&%^^^^
вывод:
SVO TRS 29481292
LJPZ DME 11113283675"""

n = int(input())
for i in range(n):
    str = input()
    if str.find('%%') == 0: 
        print(str[2:])
    elif str.find('#') == 0:
        pass
    else:
        print(str)
```

```python
"""3.От и до
Напишите программу, которая считывает список чисел, а затем выводит сумму в
диапазоне от заданного до заданного. При этом в программе не должно
использоваться обращение к элементам по индексу.
На первой строке вводится натуральное число N — количество чисел
в списке.
Далее следуют N целых чисел из списка, каждое на отдельной
строке.
Затем на отдельных строках следуют начальный (p) и конечный (q) номера
элементов списка, который нужно просуммировать. Нумерация идёт с единицы.
Гарантируется,
что 1≤ p ≤ q ≤ N.
Пример:
ввод:
5
100
222
300
444
500
2
4
вывод:
966"""

n = int(input())
values = [int(input()) for i in range(n)]
p = int(input())
q = int(input())
print()
for i in range(n - q):
    values.pop()
values.reverse()
for i in range(p - 1):
    values.pop()
print(sum(values))

```

```python
"""4 Список отличников
В школе №13 отличников по программированию награждают поездкой в Артек, но
счастливых учеников может быть больше, чем свободных мест - ваша задача определить
отличников и если их больше чем свободных мест, то отсортировать их по фамилии и
вывести тех, кто едет.
Напишите функцию Artek(s, n), которая принимает в качестве параметра строку,
содержащую имена отличников и их оценку, и число - количество свободных мест. Функция
должна печатать фамилии тех, кто едет.
Пример: 
ввод:
Сидоров 3, Смирнова 5, Самойкин 4, Емельянов 2, Кац 5, Вдовиченко 5, Северин 4,
Субботкин 5
3
вывод:
Вдовиченко, Кац, Смирнова
На вход идут фамилия с оценкой через запятую, а на следующей строке кол-во свободных
мест"""

def Artek(s, n):
    s = s.split(", ")
    s = list(filter(lambda x: int(x[len(x) - 1]) % 5 == 0, s))
    s = list(map(lambda x: x.replace(" 5", ""), s))
    s.sort()
    s = s[:n]
    s = ", ".join(s)
    return s

s = str(input())
n = int(input())
print(Artek(s, n))
```

```python
"""5 Пароль (*)
Написать программу, которая просит пользователя ввести пароль и подтвердить его (как
при обычной регистрации). Пароль должен соответствовать следующим требованиям:
1)В пароле должны быть буквы (Заглавные и прописные)
2)В пароле должны быть цифры
3)В пароле должен быть хотя бы один спец. символ @#$%&*().,-_+
4)Длина пароля не менее 8 символов
Вывод:
Если пароль содержит 7 или менее символов - программа
должна печатать "Недопустимый пароль".
Если в пароле содержатся только цифры или только буквы одного
регистра, функция должна печатать "Ненадежный пароль".
Если в пароле содержатся только буквы разных регистров или буквы
одного регистра и цифры, функция должна печатать "Слабый пароль".
Если в пароле содержатся и цифры, буквы разных регистров и спец символы, то
функция должна печатать "Надежный пароль".
Напишите функцию password_level(), которая
принимает на вход строку и определяет, является ли она надежным паролем."""

def password_level(s):
    special = "@#$%&*().,-_+"
    if len(s) < 8:
        print("Недопустимый пароль")
    elif s.isupper() or s.islower() or s.isdigit():
        print("Ненадежный пароль")
    else:
        up_mark = 0
        down_mark = 0
        digit_mark = 0
        special_mark = 0
        for i in range(len(s)):
            if s[i].isupper(): 
                up_mark = up_mark or 1
            elif s[i].islower():
                down_mark = down_mark or 1
            elif s[i].isdigit():
                digit_mark = digit_mark or 1
            elif s[i] in special:
                special_mark = special_mark or 1
        if up_mark + down_mark + digit_mark + special_mark == 4:
            print("Надежный пароль")
        else:
            print("Слабый пароль")
   


s = input("введите пароль")
password_level(s)
```
